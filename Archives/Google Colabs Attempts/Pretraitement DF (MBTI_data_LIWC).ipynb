#Data -> LIWC

```python
DATA_LIWC = pd.read_csv('/content/drive/MyDrive/LIWCdone/Data_LIWC2015 Results (TXT 2 (15307 files)).csv')
```

```python
import os
import glob
```

```python
os.chdir(r"/content/drive/MyDrive/LIWC chunks/TXT")
```

```python
files = [i for i in glob.glob("*.txt")]
```

```python
DATA = pd.DataFrame(columns=['Title','Name','Speech'])
```

```python
for file in files[:8000]:
  _cut = re.split('\s', file)
  Title = _cut[0].rstrip('.txt')
  Name = " ".join(_cut[1:]).rstrip('.txt')

  with open(file, mode='r') as f:
    d = f.read()

    DATA.loc[len(DATA.index)] = [Title, Name, d]

  del Title,Name,_cut
```

```python
for file in files[8000:]:
  _cut = re.split('\s', file)
  Title = _cut[0].rstrip('.txt')
  Name = " ".join(_cut[1:]).rstrip('.txt')

  with open(file, mode='r') as f:
    d = f.read()

    DATA.loc[len(DATA.index)] = [Title, Name, d]

  del Title,Name,_cut
```

```python
DATA.to_csv('DATA.csv', header=True, index=True)
```
